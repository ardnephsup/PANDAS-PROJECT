# groupby()

```python
#groupby()
# groupby() is one of the most important pandas functions for data aggregation and transformation.
#It allows you to group data based on a certain column or columns, and then apply aggregation
#functions like sum(), mean(), count() 

df.groupby('City')['Age'].mean()
```

```python
#adding a column named salary
df['Salary']= [50000, 60000, 75000, 52000, 45000]
df
```

```python
#on the basis of city we group by 2 columns age and salary and calculated mean

df.groupby('City')[['Age','Salary']].mean()
```

# sort_values()

**Sorting data is one of the most common operations when analyzing or exploring datasets. sort_values() allows you to sort the rows in ascending or descending order based on one or more columns.

```python
#sorted single column

sorted_df_age = df.sort_values(by='Age')
sorted_df_age
```

```python
#sorted multiple column

sorted_df_multiple = df.sort_values(by=['Age', 'Salary'], ascending=[True, False])
sorted_df_multiple
```
